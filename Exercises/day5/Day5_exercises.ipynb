# Scipy

```python
import numpy as np
from scipy import linalg 
```

```python
#a. Define a matrix A
A = np.array([[1,-2, 3],
             [4, 5, 6],
             [7, 1, 9]])
```

```python
#b. Define a vector b
b = [1,2,3]
```

```python
#c. Solve the linear system of equations A x = b
x = linalg.solve(A,b)
print(x)
```

```python
#d. Check that your solution is correct by plugging it into the equation
np.dot(A,x)
```

```python
#e. Repeat steps a-d using a random 3x3 matrix B (instead of the vector b)

#create the random b matirx
bb = np.random.random([3,3])
print(f"b matrix{bb}")
#solve the equation for x
xx = linalg.solve(A,bb)
#test the b value
np.dot(A,xx)
```

```python
#f. Solve the eigenvalue problem for the matrix A and print the eigenvalues and eigenvectors
eigen = linalg.eig(A)
print(eigenvalues)
```

```python
#g. Calculate the inverse, determinant of A

#inverse
inv = linalg.inv(A)
print(f"the inverse of A {inv}")
#determinant
det = linalg.det(A)
print(f"The determinant of A: {det}")
```

```python
#h. Calculate the norm of A with different orders

#with order of one
norm = linalg.norm(A, 1)
print(f"The norm with order of one is: {norm}")

#with order of -1
norm = linalg.norm(A, -1)
print(f"The norm with order of negative one is: {norm}")

#with order of 2
norm = linalg.norm(A, 2)
print(f"The norm with order of two is: {norm:.2f}")

#with order of None
norm = linalg.norm(A, None)
print(f"The norm with order of None is: {norm:.2f}")
```

# Statistics

a. Create a discrete random variable with poissonian distribution and plot its probability mass function (PMF), cummulative distribution function (CDF) and a histogram of 1000 random realizations of the variable

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

#create the random variable
mu = 6
v = poisson.rvs(mu=5, size = 10000)
print("First 10 values of the radom variable:")
print(v[:10])

#plotting PMF
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(v, poisson.pmf(v,mu), 'ro', ms=5, label='poisson pmf')
ax.legend()
```

```python
import seaborn as sns

#calculate and plot CDF
cdf = poisson.cdf(v,mu)

sns.kdeplot(data=cdf, cumulative=True, label = 'CDF')
plt.legend()
plt.show()
```

```python
#draw 1000 realization and plot a histogram of their values

n = 1000
realizations = np.random.choice(v, size=n, replace=True)

plt.figure(figsize=(6,4))
plt.hist(realizations, bins=np.max(realizations) - np.min(realizations), alpha=0.7, edgecolor='black')
```

b. Create a continious random variable with normal distribution and plot its probability mass function (PMF), cummulative distribution function (CDF) and a histogram of 1000 random realizations of the variable

```python
from scipy.stats import poisson
from scipy.stats import norm
import matplotlib.pyplot as plt

#create the random variable
mu = 0
sigma = 1
n = 10000

v = np.random.normal(mu, sigma, n)
print("First 10 values of the radom variable:")
print(v[:10])

#calculate PDF and CDF
pdf = norm.pdf(v, mu, sigma)
cdf = norm.cdf(v, mu, sigma)

#plotting 
fig, ax = plt.subplots(figsize=(6,4))

ax.hist(v, 30, density=True)
ax.plot(v, pdf, 'o', color = 'r', ms=1, label = 'PDF')
ax.legend()

fig, ax = plt.subplots(figsize=(6,4))
sns.kdeplot(data=cdf, cumulative=True, label = 'CDF')
ax.legend()

#draw 1000 realization and plot a histogram of their values
realizations = np.random.choice(v, size=1000, replace=True)
fig, ax = plt.subplots(figsize=(6,4))
plt.hist(realizations, bins=20, alpha=0.7, edgecolor='black')

```

c. Test if two sets of (independent) random data comes from the same distribution

```python
from scipy.stats import ttest_ind

#create two sets of independent random data
data1 = np.random.normal(loc=3, scale=1, size=100)
data2 = np.random.normal(loc=5, scale=1, size=100)

#perform the t-test
t_stat, p_value = ttest_ind(data1, data2)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

#conclusion
if p_value < 0.05:
    print("Reject the null hypothesis: The two datasets come from different distributions.")
else:
    print("Fail to reject the null hypothesis: The two datasets may come from the same distribution.")
```
